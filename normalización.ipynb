```python
import pandas as pd
import unicodedata
import re

# Función de normalización
def normaliza_lista(texto):
    if pd.isna(texto):
        return []
    lista = texto.split('|')
    lista_limpia = []
    for item in lista:
        item = item.strip().lower()
        item = ''.join(
            c for c in unicodedata.normalize('NFD', item)
            if unicodedata.category(c) != 'Mn'
        )
        item = re.sub(r'[^\w\s]', '', item)
        if item:
            lista_limpia.append(item)
    return lista_limpia

df = pd.read_csv('datos.csv')

# Normalizar autores y afiliaciones
df['autores_normalizados'] = df['autores'].apply(normaliza_lista)
df['afiliaciones_normalizadas'] = df['afiliaciones'].apply(normaliza_lista)

# Convertir las listas a texto plano separado por coma
df['autores_normalizados'] = df['autores_normalizados'].apply(lambda x: ', '.join(x))
df['afiliaciones_normalizadas'] = df['afiliaciones_normalizadas'].apply(lambda x: ', '.join(x))

# Dejar solo las columnas deseadas
df_final = df[['autores_normalizados', 'afiliaciones_normalizadas', 'tematica', 'palabras_clave', 'titulo']]

df_final.to_csv('datos_normalizados.csv', index=False)
```

```python
import pandas as pd

# Cargar el archivo original
df = pd.read_csv("datos_normalizados.csv")

# Eliminar títulos duplicados (exactamente iguales), conservar el primero
df_sin_duplicados = df.drop_duplicates(subset='titulo', keep='first')

# Reordenar columnas
df_sin_duplicados = df_sin_duplicados[[
    'titulo',
    'tematica',
    'autores_normalizados',
    'afiliaciones_normalizadas',
    'palabras_clave'
]]

# Guardar el nuevo CSV
df_sin_duplicados.to_csv("data_final.csv", index=False)
```
