```python
import os
import pdfplumber
```

```python
carpeta_pdfs = "papers"
carpeta_salida = "outputs"

os.makedirs(carpeta_salida, exist_ok=True)
```

```python
#Función para extraer texto de un PDF
def extraer_texto_pdf(ruta_pdf):
    texto_completo = ""
    with pdfplumber.open(ruta_pdf) as pdf:
        for pagina in pdf.pages:
            texto = pagina.extract_text()
            if texto:
                texto_completo += texto + "\n"
    return texto_completo

#Procesar todos los PDFs de la carpeta
for nombre_archivo in os.listdir(carpeta_pdfs):
    if nombre_archivo.lower().endswith(".pdf"):
        ruta_pdf = os.path.join(carpeta_pdfs, nombre_archivo)
        texto = extraer_texto_pdf(ruta_pdf)
        
        nombre_salida = os.path.splitext(nombre_archivo)[0] + ".txt"
        ruta_salida = os.path.join(carpeta_salida, nombre_salida)

        with open(ruta_salida, "w", encoding="utf-8") as f:
            f.write(texto)
        
        print(f"✅ Procesado: {nombre_archivo} → {ruta_salida}")
```
